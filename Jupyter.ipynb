### Text Formatting

Fomatting the text in various formats 

1. Bold 
    1. Using double underscore before and after text modified
          1.  __Bold__
2. Italics 
    1. Using single underscore befor and after text
        1. ___Italics___
3. Color
   1. Using the font tag and specifying the color like a CSS property
       1. <font color=red>_Aashish_</font>
            <font color=green>Aashish</font>
            <font color=blue>Aashish</font>
            <font color=pink>Aashish</font>
            <font color=yellow>__Aashish__</font>
       
4. Line Break
    1. Using the __<br__> tag for the same
    
5. Indentation 
    1. Using the '>' symbol before the text to be indented
        1. > I am indented
        
6. Mathematical Expression
    1. Using dollar symbol before and after expression
        1. $ x =  y+ z$
7. Using the text in highlighted format
    `Highlighted Text` 
8. Adding Links 
    1. [DisplayName](Reference)
9. Horizontal Line Separator
   1.  Something 
       *** 
       Is written
